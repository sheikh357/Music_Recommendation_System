```python
import pandas as pd
```

```python
df = pd.read_csv('/home/maemoon/Documents/BDA_Project/fma_metadata/raw_tracks.csv')
df2=pd.read_csv('extracted_features.csv')
```

```python
column_names = df2.columns.tolist()
print(column_names)
```

```python
newdf = pd.DataFrame()
newdf['track_id'] = df2['file']
```

```python
newdf['track_id'] = newdf['track_id'].str.lstrip('0').str.rstrip('.mp3')
```

```python
# Convert 'track_id' to string in both DataFrames
df['track_id'] = df['track_id'].astype(str)
newdf['track_id'] = newdf['track_id'].astype(str)

# Merge the DataFrames on 'track_id'
merged_df = pd.merge(df, newdf, on='track_id', how='inner')

# Create a new DataFrame with both 'track_id' and 'artist_name'
artist_names_df = merged_df[['track_id', 'artist_name','track_title','track_genres']].copy()
```

```python
artist_names_df
```

```python
df2['file'] = df2['file'].str.lstrip('0').str.rstrip('.mp3')
```

```python
df2
```

```python
df2.rename(columns={'file': 'track_id'}, inplace=True)

# Merge the two DataFrames based on the common column
merged_df = pd.merge(artist_names_df, df2, on='track_id', how='inner')

# Display the merged DataFrame
print(merged_df)
```
